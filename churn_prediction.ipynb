```python
plt.bar(df['Customer Status'].value_counts().index, df['Customer Status'].value_counts().values)
plt.xlabel('Customer Status')
plt.ylabel('Count')
plt.title('Distribution of Customer Status')
plt.show()
```

```python
plt.bar(df['Churn Category'].value_counts().index, df['Churn Category'].value_counts().values)
plt.xlabel('Churn Category')
plt.ylabel('Count')
plt.title('Distribution of Churn Category')
plt.show()
```

```python
# Create a cross-tabulation of Customer Status and Gender
status_gender_counts = df.groupby(['Customer Status', 'Gender']).size().unstack()

# Plotting the grouped bar chart
status_gender_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Customer Status')
plt.ylabel('Count')
plt.title('Distribution of Gender by Customer Status')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
plt.bar(df['Internet Type'].value_counts().index, df['Internet Type'].value_counts().values)
plt.xlabel('Internet Type')
plt.ylabel('Count')
plt.title('Distribution of Internet Type')
plt.show()
```

```python
plt.bar(df['Contract'].value_counts().index, df['Contract'].value_counts().values)
plt.xlabel('Contract')
plt.ylabel('Count')
plt.title('Distribution of Contract')
plt.figure(figsize=(10, 6))
plt.show()
```
